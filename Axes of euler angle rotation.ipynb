```julia
using Plots
pyplot(
    legend=false,
    xlims=(-1.0,1.0),ylims=(-1.0,1.0),zlims=(-1.0,1.0),
    xlab="x",ylab="z",zlab="y",
    ticks=nothing
)
function plot_axes(cmrx,cmry)
    p=plot(θ->-sin(θ),θ->0,θ->-cos(θ),0,2π,camera=(cmrx,cmry))
    for θ = 0:π/16:2π
        rx=-1+cos(θ)-sin(θ)
        ry=1+cos(θ)-sin(θ)
        rz=-1+cos(θ)+sin(θ)
        r=√(rx^2+ry^2+rz^2)
        plot!(p,[0.0,rx/r],[0.0,rz/r],[0.0,-ry/r],
              color=:blue,lw=ifelse(θ<π/2,3,1),camera=(cmrx,cmry))
    end
    p
end
pxy = plot_axes(0,0)
pzy = plot_axes(90,0)
plot(
    pxy,
    pzy,
    layout=2,size=(1000,500)
)
```

```julia
using Plots
gr(
    xlims=(0.0,2π),ylims=(0.0,3π/2),
    xlab="θ",ylab="φ",
    ticks=0:π/4:2π,
    size=(577,433),
    legend=false
)
f(θ) = 2acos((1+cos(θ)+sin(θ))/(2*√2))
plot(θ->θ, f, 0.0, π/2, color=:red, lw=3)
plot!(θ->θ, f, π/2, 2π, color=:red, lw=1)
```
